```python
df_clean = df[(df['Quantity'] >= 1) & (df['UnitPrice'] > 0)]
```

```python
df_clean.shape
```

```python
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'], dayfirst=True)
```

```python
df_clean.info()
```

```python
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['UnitPrice']
```

```python
df_clean.head()
```

```python
df_customers = df_clean.dropna(subset=['CustomerID'])
```

```python
df_customers.shape
```

```python
df_clean.to_csv("clean_retail_data.csv", index = False)
```

```python
df_customers.to_csv("customers_data.csv", index = False)
```
